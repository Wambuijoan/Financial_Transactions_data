# **Working with a JSON File from GitHub**

---



```python
import requests
import pandas as pd
```

```python
# Fetching and reading JSON from the Github URL
financial_data_url = 'https://raw.githubusercontent.com/Wambuijoan/Financial_Transactions_data/refs/heads/main/Transactions_data.json'

data = requests.get(financial_data_url)

# Checing if the request has succeeded:
print(data.status_code)
```

```python
# Assigning a variable name to the data
financial_data = data.json()
```

```python
print(type(financial_data))
```

```python
# Using Pandas to convert the JSON file into a DataFrame:
financial_data_df = pd.DataFrame(financial_data)
financial_data_df.head()
```

```python
# Display top records in the DataFrame:
print(financial_data_df.head())
```

```python
# Saving as CSV

financial_data_df.to_csv('financial_data.csv', index=False)

print("Financial data CSV has been saved.")
```

# **Working with DummyJSON API Endpoints**

---


Dealing with Products Dummy JSON API Endpoints:

```python
products_data_url = 'https://dummyjson.com/products'

products_data = requests.get(products_data_url)

# Checking if the request has succeeded:
products_data.status_code
```

```python
products_data = products_data.json()

print(type(products_data))
```

```python
products_data_df = pd.DataFrame(products_data)
```

```python
print(products_data_df.head())
```
